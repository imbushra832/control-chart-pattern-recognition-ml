```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import PipelineM
from sklearn.neural_network import MLPClassifier

# Helper function to load and preprocess the data
def load_data(file_path):
    data = np.loadtxt(file_path)
    X = data[:, 0:10]  # First 10 columns are inputs
    y = data[:, 10:17]  # Next 7 columns are outputs
    return X, y

# Load training and testing data
X_train, y_train = load_data("SPC-Training.txt")
X_test, y_test = load_data("SPC-Testing.txt")

# Convert one-hot encoded outputs to class labels for certain algorithms
y_train_labels = np.argmax(y_train, axis=1)
y_test_labels = np.argmax(y_test, axis=1)

# Normalize the input data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Print dataset information
print(f"Training set: {X_train.shape[0]} samples")
print(f"Testing set: {X_test.shape[0]} samples")
print(f"Class distribution in training set:")
for i in range(7):
    count = np.sum(y_train_labels == i)
    print(f"  Class {i}: {count} samples ({count/len(y_train_labels)*100:.1f}%)")

# Function to plot training history
def plot_history(history):
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='lower right')
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper right')
    
    plt.tight_layout()
    plt.show()

# Function to evaluate and print results
def evaluate_model(y_true, y_pred, model_name):
    print(f"\n--- {model_name} Results ---")
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, 
                               target_names=['In Control', 'Rule 1', 'Rule 2', 'Rule 3', 'Rule 4', 'Rule 5', 'Rule 6']))
    
    # Plot confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.colorbar()
    tick_marks = np.arange(7)
    plt.xticks(tick_marks, ['In Control', 'Rule 1', 'Rule 2', 'Rule 3', 'Rule 4', 'Rule 5', 'Rule 6'], rotation=45)
    plt.yticks(tick_marks, ['In Control', 'Rule 1', 'Rule 2', 'Rule 3', 'Rule 4', 'Rule 5', 'Rule 6'])
    
    # Add text annotations
    thresh = cm.max() / 2
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, format(cm[i, j], 'd'),
                    horizontalalignment="center",
                    color="white" if cm[i, j] > thresh else "black")
    
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
    
    return accuracy_score(y_true, y_pred)

# 1. MLP Implementation
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

def train_mlp():
    print("\n=== Training Optimized MLP Network ===")
    
    # Calculate class weights
    class_weights = compute_class_weight('balanced', classes=np.unique(y_train_labels), y=y_train_labels)
    class_weights = {i: class_weights[i] for i in range(len(class_weights))}

    # Create the optimized model
    model = Sequential([
        Dense(128, kernel_regularizer=l2(0.01), input_shape=(10,)),
        BatchNormalization(),
        LeakyReLU(alpha=0.1),
        Dropout(0.3),
        
        Dense(64, kernel_regularizer=l2(0.01)),
        BatchNormalization(),
        LeakyReLU(alpha=0.1),
        Dropout(0.3),
        
        Dense(32, kernel_regularizer=l2(0.01)),
        BatchNormalization(),
        LeakyReLU(alpha=0.1),
        Dropout(0.2),
        
        Dense(7, activation='softmax')
    ])
    
    # Compile the model
    optimizer = Adam(learning_rate=0.0001)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    
    # Callbacks
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001)
    
    # Train the model
    history = model.fit(
        X_train_scaled, y_train,
        epochs=200,
        batch_size=32,
        validation_split=0.2,
        callbacks=[early_stopping, reduce_lr],
        class_weight=class_weights
    )
    
    # Plot training history
    plot_history(history)
    
    # Predictions
    y_pred_proba = model.predict(X_test_scaled)
    y_pred = np.argmax(y_pred_proba, axis=1)
    
    # Evaluate
    mlp_acc = evaluate_model(y_test_labels, y_pred, "MLP")
    
    return model, mlp_acc
# 2. RBF Implementation using scikit-learn's MLPClassifier with custom centers
def train_rbf():
    print("\n=== Training RBF Network ===")
    
    # Set up parameter grid for grid search
    param_grid = {
        'hidden_layer_sizes': [(50,), (100,), (150,)],
        'activation': ['tanh'],  # Using tanh to mimic RBF behavior
        'alpha': [0.0001, 0.001, 0.01],
        'learning_rate_init': [0.001, 0.01]
    }
    
    # Create the MLPClassifier
    mlp = MLPClassifier(max_iter=500, early_stopping=True, verbose=1)
    
    # Create grid search
    grid_search = GridSearchCV(
        mlp, param_grid, cv=3, scoring='accuracy', verbose=1, n_jobs=-1
    )
    
    # Fit the grid search
    grid_search.fit(X_train_scaled, y_train_labels)
    
    # Get the best model
    best_rbf = grid_search.best_estimator_
    print(f"Best parameters: {grid_search.best_params_}")
    
    # Make predictions
    y_pred = best_rbf.predict(X_test_scaled)
    
    # Evaluate the model
    rbf_acc = evaluate_model(y_test_labels, y_pred, "RBF")
    
    return best_rbf, rbf_acc

#3. SVM Implementation
def train_svm():
    print("\n=== Training SVM ===")
    
    # Create pipeline with scaler and SVM
    svm_pipeline = Pipeline([
        ('svm', OneVsRestClassifier(SVC(probability=True), n_jobs=1))  # Avoid excessive parallelization
    ])
    
    # Define reduced parameter grid
    param_grid = {
        'svm__estimator__C': [0.1, 1, 10],  # Reduced
        'svm__estimator__gamma': ['scale', 'auto', 0.1],  # Reduced
        'svm__estimator__kernel': ['rbf']  # Removed 'poly' for now
    }
    
    # Perform grid search
    grid_search = GridSearchCV(
        svm_pipeline, param_grid, cv=3, scoring='accuracy', verbose=1, n_jobs=4  # Limit parallel jobs
    )
    
    print("Starting Grid Search for SVM...")
    try:
        grid_search.fit(X_train_scaled, y_train_labels)
        print(f"Best parameters: {grid_search.best_params_}")
        
        # Get best model
        best_svm = grid_search.best_estimator_
        
        # Make predictions
        y_pred = best_svm.predict(X_test_scaled)
        
        # Evaluate the model
        svm_acc = evaluate_model(y_test_labels, y_pred, "SVM")
        
        return best_svm, svm_acc
    
    except Exception as e:
        print(f"Error during SVM training: {e}")
        return None, 0

# 5. Compare all models
def compare_models(accuracies):
    models = list(accuracies.keys())
    accs = list(accuracies.values())
    
    plt.figure(figsize=(10, 6))
    plt.bar(models, accs, color=['blue', 'green', 'red', 'purple'])
    plt.xlabel('Model')
    plt.ylabel('Accuracy')
    plt.title('Model Comparison')
    plt.ylim(0, 1.0)
    
    # Add accuracy values on top of bars
    for i, v in enumerate(accs):
        plt.text(i, v + 0.01, f'{v:.4f}', ha='center')
    
    plt.tight_layout()
    plt.show()


def visualize_sample_patterns():
    # Plot examples of each class
    plt.figure(figsize=(20, 12))
    class_names = ['In Control', 'Rule 1', 'Rule 2', 'Rule 3', 'Rule 4', 'Rule 5', 'Rule 6']
    
    for class_idx in range(7):
        # Get samples for this class
        class_samples = X_train[y_train_labels == class_idx]
        
        # Plot up to 3 examples per class
        for i in range(min(3, len(class_samples))):
            plt.subplot(7, 3, class_idx*3 + i + 1)
            plt.plot(class_samples[i], '-o')
            plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)
            plt.axhline(y=1, color='g', linestyle='--', alpha=0.3)
            plt.axhline(y=-1, color='g', linestyle='--', alpha=0.3)
            plt.axhline(y=2, color='y', linestyle='--', alpha=0.3)
            plt.axhline(y=-2, color='y', linestyle='--', alpha=0.3)
            plt.axhline(y=3, color='m', linestyle='--', alpha=0.3)
            plt.axhline(y=-3, color='m', linestyle='--', alpha=0.3)
            plt.title(f"{class_names[class_idx]} Pattern {i+1}")
            plt.ylim(-4, 4)
            plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('spc_pattern_examples.png')
    plt.show()

def visualize_decision_boundaries(model, centers=None, width=None, model_type='mlp'):
    """
    Visualize decision boundaries in 2D using PCA
    Note: This is a simplified visualization that projects to 2D
    """
    from sklearn.decomposition import PCA
    
    # Project data to 2D for visualization
    pca = PCA(n_components=2)
    X_train_2d = pca.fit_transform(X_train_scaled)
    X_test_2d = pca.transform(X_test_scaled)
    
    # Create a mesh grid
    h = 0.02  # Step size
    x_min, x_max = X_train_2d[:, 0].min() - 1, X_train_2d[:, 0].max() + 1
    y_min, y_max = X_train_2d[:, 1].min() - 1, X_train_2d[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    # Create mesh points and transform back to original feature space
    mesh_points_2d = np.c_[xx.ravel(), yy.ravel()]
    mesh_points = pca.inverse_transform(mesh_points_2d)
    
    # Make predictions
    if model_type == 'custom_rbf':
        # For custom RBF, we need to apply RBF transform first
        from scipy.spatial.distance import cdist
        dists = cdist(mesh_points, centers)
        mesh_points_rbf = np.exp(-(dists ** 2) / (2 * width ** 2))
        Z = model.predict(mesh_points_rbf)
    elif model_type == 'keras':
        # For Keras models
        Z = np.argmax(model.predict(mesh_points), axis=1)
    else:
        # For scikit-learn models
        Z = model.predict(mesh_points)
    
    # Reshape result
    Z = Z.reshape(xx.shape)
    
    # Plot decision boundary
    plt.figure(figsize=(12, 10))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Spectral)
    
    # Plot training and testing points
    class_names = ['In Control', 'Rule 1', 'Rule 2', 'Rule 3', 'Rule 4', 'Rule 5', 'Rule 6']
    colors = plt.cm.rainbow(np.linspace(0, 1, 7))
    
    # Plot training data
    for i in range(7):
        idx = y_train_labels == i
        plt.scatter(X_train_2d[idx, 0], X_train_2d[idx, 1], c=[colors[i]], 
                   label=f'Train: {class_names[i]}', alpha=0.6, edgecolors='k')
    
    # If centers are provided (for custom RBF), plot them
    if centers is not None:
        centers_2d = pca.transform(centers)
        plt.scatter(centers_2d[:, 0], centers_2d[:, 1], c='k', marker='*', s=150,
                   alpha=0.8, edgecolors='w', label='RBF Centers')
    
    plt.title(f'Decision Boundaries ({model_type.upper()})')
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
    plt.tight_layout()
    plt.savefig(f'decision_boundaries_{model_type}.png')
    plt.show()

def visualize_feature_importance(best_model, model_type='svm'):
    """Visualize feature importance for the different models"""
    plt.figure(figsize=(12, 6))
    
    if model_type == 'mlp':
        # For MLP, visualize weights of the first layer
        weights = np.abs(best_model.get_weights()[0])
        importance = np.sum(weights, axis=1)
        
    elif model_type == 'svm':
        # For SVM, use coefficients if linear kernel, otherwise return
        if best_model.named_steps['svm'].estimator.kernel == 'linear':
            importance = np.abs(best_model.named_steps['svm'].estimator.coef_).sum(axis=0)
        else:
            print("Feature importance visualization not available for non-linear SVM kernels")
            return
            
    elif model_type == 'custom_rbf':
        # For Custom RBF, use logistic regression coefficients
        importance = np.abs(best_model.coef_).sum(axis=0)
        # Since feature importance is for RBF features, not original features
        plt.bar(range(len(importance)), importance)
        plt.title(f'RBF Feature Importance ({model_type.upper()})')
        plt.xlabel('RBF Center Index')
        plt.ylabel('Importance')
        plt.tight_layout()
        plt.savefig(f'feature_importance_{model_type}.png')
        plt.show()
        return
    
    # Plot feature importance
    feature_names = [f'Feature {i+1}' for i in range(10)]
    plt.bar(feature_names, importance)
    plt.title(f'Feature Importance ({model_type.upper()})')
    plt.xlabel('Feature')
    plt.ylabel('Importance')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f'feature_importance_{model_type}.png')
    plt.show()

def analyze_misclassifications(model, X_data, y_true, model_type='mlp'):
    """Analyze which patterns are most often misclassified"""
    if model_type == 'keras':
        y_pred_proba = model.predict(X_data)
        y_pred = np.argmax(y_pred_proba, axis=1)
    elif model_type == 'custom_rbf':
        from scipy.spatial.distance import cdist
        dists = cdist(X_data, centers)
        X_rbf = np.exp(-(dists ** 2) / (2 * width ** 2))
        y_pred = model.predict(X_rbf)
    else:
        y_pred = model.predict(X_data)
    
    # Get misclassified instances
    misclassified = y_pred != y_true
    
    # Count misclassifications by true class
    class_names = ['In Control', 'Rule 1', 'Rule 2', 'Rule 3', 'Rule 4', 'Rule 5', 'Rule 6']
    misclass_counts = np.zeros(7)
    class_counts = np.zeros(7)
    
    for i in range(7):
        class_idx = y_true == i
        class_counts[i] = np.sum(class_idx)
        misclass_counts[i] = np.sum(misclassified & class_idx)
    
    # Calculate error rates
    error_rates = misclass_counts / class_counts
    
    # Plot error rates
    plt.figure(figsize=(12, 6))
    plt.bar(class_names, error_rates)
    plt.title(f'Misclassification Rates by Pattern Type ({model_type.upper()})')
    plt.xlabel('Pattern Type')
    plt.ylabel('Error Rate')
    plt.xticks(rotation=45)
    
    # Add percentage labels
    for i, v in enumerate(error_rates):
        plt.text(i, v + 0.01, f'{v:.2%}', ha='center')
    
    plt.tight_layout()
    plt.savefig(f'misclassification_rates_{model_type}.png')
    plt.show()
    
    # Find the most confusing pairs of classes
    cm = confusion_matrix(y_true, y_pred)
    
    # Set diagonal elements to 0 to focus on misclassifications
    np.fill_diagonal(cm, 0)
    
    # Get the top confusion pairs
    confusion_pairs = []
    for i in range(7):
        for j in range(7):
            if i != j:
                confusion_pairs.append((i, j, cm[i, j]))
    
    # Sort confusion pairs by count
    confusion_pairs.sort(key=lambda x: x[2], reverse=True)
    
    # Print top confusion pairs
    print("\nTop Confusion Pairs:")
    for i, j, count in confusion_pairs[:5]:
        print(f"{class_names[i]} mistaken as {class_names[j]}: {count} times ({count/class_counts[i]:.2%} of class {i})")
    
    return misclassified, error_rates

def create_ensemble_model():
    """Create an ensemble of the trained models for better performance"""
    print("\n=== Creating Ensemble Model ===")
    
    # Train individual models
    mlp_model, _ = train_mlp()
    rbf_model, _ = train_rbf()
    custom_rbf_model, _, centers, width = custom_rbf()
    svm_model, _ = train_svm()
    
    # Function to get predictions from each model
    def get_ensemble_predictions(X):
        # MLP predictions
        mlp_proba = mlp_model.predict(X)
        mlp_pred = np.argmax(mlp_proba, axis=1)
        
        # RBF predictions
        rbf_pred = rbf_model.predict(X)
        
        # Custom RBF predictions
        from scipy.spatial.distance import cdist
        dists = cdist(X, centers)
        X_rbf = np.exp(-(dists ** 2) / (2 * width ** 2))
        custom_rbf_pred = custom_rbf_model.predict(X_rbf)
        
        # SVM predictions
        svm_pred = svm_model.predict(X)
        
        # Combine predictions using voting
        pred_matrix = np.column_stack((mlp_pred, rbf_pred, custom_rbf_pred, svm_pred))
        ensemble_pred = np.zeros(len(X), dtype=int)
        
        for i in range(len(X)):
            # Get most common prediction for this instance
            vals, counts = np.unique(pred_matrix[i], return_counts=True)
            ensemble_pred[i] = vals[np.argmax(counts)]
        
        return ensemble_pred
    
    # Evaluate ensemble model
    y_pred_ensemble = get_ensemble_predictions(X_test_scaled)
    ensemble_acc = evaluate_model(y_test_labels, y_pred_ensemble, "Ensemble")
    
    # Compare with individual models
    accuracies = {
        'MLP': evaluate_model(y_test_labels, np.argmax(mlp_model.predict(X_test_scaled), axis=1), "MLP (Again)"),
        'RBF': evaluate_model(y_test_labels, rbf_model.predict(X_test_scaled), "RBF (Again)"),
        'Custom RBF': evaluate_model(y_test_labels, custom_rbf_model.predict(np.exp(-(cdist(X_test_scaled, centers) ** 2) / (2 * width ** 2))), "Custom RBF (Again)"),
        'SVM': evaluate_model(y_test_labels, svm_model.predict(X_test_scaled), "SVM (Again)"),
        'Ensemble': ensemble_acc
    }
    
    # Compare models
    compare_models(accuracies)
    
    return get_ensemble_predictions, accuracies
```

```python
# Train and evaluate SVM
_, svm_acc = train_svm()
accuracies['SVM'] = svm_acc
```

```python
# Compare all models
compare_models(accuracies)
```
